```python
import numpy as np
import pandas as pd

from keras.models import Sequential
from keras.layers.core import Dense
from keras.utils import np_utils
from keras import optimizers

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg' 

from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
```

```python
df_train = pd.read_csv("nums_train.csv")
X_train, Y_train = df_train.values[:, 1:], df_train.values[:, 0]

df_test = pd.read_csv("nums_test.csv")
X_test, Y_test = df_test.values[:, 1:], df_test.values[:, 0]
```

```python
Y_train = np_utils.to_categorical(Y_train)
Y_test = np_utils.to_categorical(Y_test)
```

```python
model_sgd = Sequential()

model_sgd.add(Dense(output_dim = 10, activation = 'sigmoid'))
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=False)
model_sgd.compile(optimizer= sgd, loss='categorical_crossentropy', metrics=['accuracy'])
history_sgd = model_sgd.fit(X_train, Y_train, validation_data = (X_test, Y_test), epochs = 600);
```

```python
model_sgd_nesterov = Sequential()

model_sgd_nesterov.add(Dense(output_dim = 10, activation = 'sigmoid'))
sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model_sgd_nesterov.compile(optimizer= sgd, loss='categorical_crossentropy', metrics=['accuracy'])
history_sgd_nesterov = model_sgd_nesterov.fit(X_train, Y_train, validation_data = (X_test, Y_test), epochs = 600);
```

```python
model_adam = Sequential()

model_adam.add(Dense(output_dim = 10, activation = 'sigmoid'))
model_adam.compile(optimizer= 'adam', loss='categorical_crossentropy', metrics=['accuracy'])
history_adam = model_adam.fit(X_train, Y_train, validation_data = (X_test, Y_test), epochs = 600);
```

```python
model_adagrad = Sequential()

model_adagrad.add(Dense(output_dim = 10, activation = 'sigmoid'))
model_adagrad.compile(optimizer= 'adagrad', loss='categorical_crossentropy', metrics=['accuracy'])
history_adagrad = model_adagrad.fit(X_train, Y_train, validation_data = (X_test, Y_test), epochs = 600);
```

```python
def sequence(x):
    return 4*x*(1-x)
```

```python
def generate_sequence(x0, n):
    x = [x0]
    for i in range(1, n):
        x.append(sequence(x[-1]))
    return x
```

```python
S = generate_sequence(0.3, 1000000)

X = S[:-1]
Y = S[1:]
```

```python
v = 0.5
```

```python
w_0 = np.ones(5)
w = np.random.rand(5)
b = np.ones(5)
```

```python
for t in range(0, 10000):
    x_0 = X[t]*w_0
    e = Y[t]
    
    z = np.abs(x_0-w)
    x = np.exp(-b*z**2)
    y = np.sum(x*w)

    for k in range(5):
        b[k] = b[k] + v*(y-e)*x[k]*w[k]*b[k]
        w[k] = w[k] - v*(y-e)*x[k]
    
```
